## Create LTI v1 token

```python
import jwt
import time
import uuid
import datetime

CANVAS_IP = "35.227.159.246"

def encode_lti1_token(id, secret):
    payload = {
        "sub": id,
        "aud": "http://{}/api/lti/accounts/1/authorize".format(CANVAS_IP),
        "exp": int(time.time() + 60), 
        "iat": int(time.time()),  
        "jti": str(uuid.uuid4())  
        }
    return jwt.encode(
        payload, 
        secret, 
        algorithm='HS256')
```
